```python
import numpy as np
import requests
from bs4 import BeautifulSoup
import time
import json
```

#### Make Dataset with Wikipedia: USC - BS4 Crawling

```python
class PageRankCrawlingBot:
    def __init__(self, wikipedia_link_main):
        self.mainlink_name = wikipedia_link_main
        self.wiki_links = {}
        self.visited = set()

    def get_wiki_links(self,url,depth=1, max_depth=2):
        # exception logic to recursive function
        mainlink_name = url.split('/')[-1]
        if depth > max_depth or mainlink_name in self.visited:
            return
        self.visited.add(mainlink_name)
        #1. set up agent for wikipedia crawler requirements
        header = {
            'User-Agent': 'PageRankStudyBot/1.0 (kimchany@usc.edu) Based on requests/2.31.0'
        }
        #2. get response and soup by bs4
        response = requests.get(url, headers = header)
        soup = BeautifulSoup(response.text, 'html.parser')
        
        #3. make dictionary data structure to save related links
        self.wiki_links[mainlink_name] = {}
        contents = soup.find('div', {'id': 'mw-content-text'})
        #4. exception: remove useless links
        for unwanted in contents.find_all(['table', 'div'], class_=['infobox', 'navbox', 'reflist']):
            unwanted.decompose()
        #5. get all links related to mainlink
        for a_tag in contents.find_all('a', href=True):
            href = a_tag['href']
            title = a_tag.get('title') or a_tag.get_text().strip()
            if href.startswith('/wiki/') and ':' not in href:
                full_url = f"https://en.wikipedia.org{href}"
                self.wiki_links[mainlink_name][full_url] = title
        next_links = list(self.wiki_links[mainlink_name].keys())
        
        #6. recursive function to get all links - depth 3
        for next_url in next_links:
            time.sleep(0.2) 
            self.get_wiki_links(next_url, depth + 1, max_depth)
        
```

```python
# main logic
if __name__ == '__main__':
    usc_main_link = 'https://en.wikipedia.org/wiki/University_of_Southern_California'
    bot = PageRankCrawlingBot(usc_main_link)
    link_related = bot.get_wiki_links(bot.mainlink_name)
```
