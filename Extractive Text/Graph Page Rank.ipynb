```python
import numpy as np
import requests
from bs4 import BeautifulSoup
import time
```

#### Make Dataset with Wikipedia: USC - BS4 Crawling

```python
#1. USC Mainpage page link
wikipedia_link_main = 'https://en.wikipedia.org/wiki/University_of_Southern_California'
#2. set up agent for requests
header = {
    'User-Agent': 'PageRankStudyBot/1.0 (kimchany@usc.edu) Based on requests/2.31.0'
}

response = requests.get(wikipedia_link_main, headers = header)
soup = BeautifulSoup(response.text, 'html.parser')
mainlink_name = wikipedia_link_main.split('/')[-1]
wiki_links[mainlink_name] = {}
contents = soup.find('div', {'id': 'mw-content-text'})
for unwanted in contents.find_all(['table', 'div'], class_=['infobox', 'navbox', 'reflist']):
    unwanted.decompose()
for a_tag in contents.find_all('a', href=True):
    href = a_tag['href']
    title = a_tag.get('title') or a_tag.get_text().strip()
    if href.startswith('/wiki/') and ':' not in href:
        full_url = f"https://en.wikipedia.org{href}"
        wiki_links[mainlink_name][full_url] = title
        
```
